# Distribución de datos en conjunto de datos sobre accidentes automovilísticos

Analizaremos un conjunto de datos de accidentes automovilísticos utilizando Python y las bibliotecas matplotlib y seaborn para visualizar datos en histogramas.

El conjunto de datos sobre accidentes automovilísticos proporciona una recopilación detallada de información relacionada con factores comunes que influyen en los accidentes de tráfico, como la gravedad de la colisión, las condiciones climáticas, los tipos de carreteras y los elementos contribuyentes, ofreciendo información valiosa para el análisis y la mejora de las medidas generales de seguridad vial.

Este conjunto de datos se obtuvo de la plataforma [kaggle](https://www.kaggle.com/datasets/jacksondivakarr/car-crash-dataset) su autor es JACKSON DIVAKAR R.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('../../Data/new dataset.xlsx')
df.head()
```

1. Hacemos una configuración de estilo visual de gráficos de seaborn

```python
sns.set(style="whitegrid")
```

## Desarrollo de histogramas sobre los accidentes automovilisticos

1. Crear un histograma para la distribución de accidentes por año

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Year'], bins=10, kde=False, color='skyblue')
plt.title('Distribución de Accidentes por Año')
plt.xlabel('Año')
plt.ylabel('Número de Accidentes')
plt.show()
```

Vamos a profundizar aun mas en estos datos.

2. Crear un histograma para la distribución de accidentes por mes

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Month'], bins=12, kde=False, color='salmon')
plt.title('Distribución de Accidentes por Mes')
plt.xlabel('Mes')
plt.ylabel('Número de Accidentes')
plt.show()
```

Podemos observar que uno de los meses donde mas se registran insidentes es en el mes de Octubre.

3. Ahora vamos a crear un histograma para la distribución de accidentes por día de la semana

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Day'], bins=7, kde=False, color='lightgreen')
plt.title('Distribución de Accidentes por Día de la Semana')
plt.xlabel('Día de la Semana')
plt.ylabel('Número de Accidentes')
plt.show()
```

Para iterpretar este grafico se puede observar que en el dia donde es mas frecuente el numero de accidentes sucede en los dias viernes de todo el año.

4. Vamos a crear un histograma donde visulicemos los accidentes durante el Día vs. la Noche

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Hour'], kde=False, palette=['#0a4a7c'])
plt.title('Accidentes durante el Día vs. la Noche')
plt.xlabel('Hora')
plt.ylabel('Número de Accidentes')
plt.show()
```

Podemos observar que los accidentes son mas frecuentes entre las 16:00 y las 17:00 horas

5. Ahora oservemos los tipos de coliciones registrados

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Collision Type'], kde=False)
plt.title('Tipos de Colisiones')
plt.xlabel('Tipo de colision')
plt.ylabel('Número de Accidentes')
plt.show()
```

En esto podemos observar que los tipos de colision que mas suceden son entre 2 vehiculos.

Estos histogramas te proporcionarán información visual sobre la distribución de accidentes a lo largo de los años, meses, días de la semana, horas del día y tipos de colisión.
