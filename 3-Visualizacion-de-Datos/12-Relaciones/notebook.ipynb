## Gr치ficos de dos l칤neas
Prueba con un gr치fico multil칤nea superponiendo dos gr치ficos de l칤neas uno encima del otro, utilizando el 'despine' de Seaborn para eliminar sus espinas superiores y derechas, y utilizando ax.twinx derivado de Matplotlib. Twinx permite que un gr치fico comparta el eje x y muestre dos ejes y. As칤, mostrar el rendimiento por colonia y el n칰mero de colonias, superpuestos:

```python
fig, ax = plt.subplots(figsize=(12,6))
lineplot = sns.lineplot(x=honey['year'], y=honey['numcol'], data=honey, 
                        label = 'Number of bee colonies', legend=False)
sns.despine()
plt.ylabel('# colonies')
plt.title('Honey Production Year over Year');

ax2 = ax.twinx()
lineplot2 = sns.lineplot(x=honey['year'], y=honey['yieldpercol'], ax=ax2, color="r", 
                         label ='Yield per colony', legend=False) 
sns.despine(right=False)
plt.ylabel('colony yield')
ax.figure.legend();
```

Aunque no hay nada que salte a la vista en torno al a침o 2003, nos permite terminar esta lecci칩n con una nota un poco m치s alegre: aunque en general hay un n칰mero de colonias en descenso, el n칰mero de colonias se est치 estabilizando aunque su rendimiento por colonia est칠 disminuyendo.
